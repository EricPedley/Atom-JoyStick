```python
import pandas as pd
from matplotlib import pyplot as plt

data_file = 'flight_data/20251007_202504.csv'

df = pd.read_csv(data_file)
print(df.head())
```

```python
accel_z = df['accelZ']

plt.plot(accel_z)
plt.title('Acceleration Z over Time')
plt.xlabel('Time (samples)')
plt.ylabel('Acceleration Z (m/s²)')
plt.grid()
plt.show()
```

```python
frontleft_duty = df['frontLeft_motor_duty']
frontright_duty = df['frontRight_motor_duty']
rearleft_duty = df['rearLeft_motor_duty']
rearright_duty = df['rearRight_motor_duty']
plt.figure()
plt.plot(frontleft_duty, label='Front Left')
plt.plot(frontright_duty, label='Front Right')
plt.plot(rearleft_duty, label='Rear Left')
plt.plot(rearright_duty, label='Rear Right')
plt.title('Motor Duty Cycles over Time')
plt.xlabel('Time (samples)')
plt.ylabel('Duty Cycle (%)')
plt.legend()
plt.grid()
plt.show()
```
